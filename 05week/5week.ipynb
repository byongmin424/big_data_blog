# while
- while 문은 반복 횟수가 정해져 있지 않을 때 사용하는 반복문
- 조건을 만족히자 않으면 무조건 반복 수행함
- 조건이 while True:인 경우 항상 참이 되므로 무한 반복하게 됨. 이경우 강제 종료
- break 문을 사용해 반복문을 빠져나올 수 있도록 함
#####
- while 문의 조건이 만족하는 동안은 while 이하의 명령 반복 실행
- 들여쓰기 주의

```python
# while
i = 0

while i <= 10:
    print(i)
    i += 1
```

```python
for i in range(0,10,4):
    if i == 4:
        continue
    print('sonny')
    
    # break와 continue 문제 나오면 맞출 줄 알아야한다.
```

```python
# for in [] 이용해서 0~10까지 출력 (print)

for i in range(0,11,1):
    print(i)
```

```python
# 2단 출력

for i in range(1,10):
    print('2*',i,'=',i*2)
```

```python
# 구구단 만들기
for i in range(2,10):
    for j in range(1,10):
        print(i,'*',j,'=',i*j)
```

# function

### 개념
    어떤 작업을 하기 위해 미리 지정된 명령문
    
    동일하게 반복해야 하는 작업을 함수로 만들어 사용하면 편리함

    파이썬에서 제공하는 내장 함수를 사용하거나 또는 사용자가 정의해서 사용 가능

### 형태
    def 예약어 사용

    함수명 다음에는 반드시 괄호와 콜론이 필요

    괄호안 인자값을 쉼표로 구분하여 입력

    def(인자, 인자):
        return 반환값

```python
def u_sum(a,b,c): # 인자값 abc
    var_sum = a+b+c # 함수 정의
    return var_sum # 반환

i = 10
j = 20
k = i*j

var = u_sum(i,j,k) #정의된 함수를 불러 var에 저장
print(var)

# 엄청나게 중요하다. 엄청나게 중요하다. 함수의 정의는 많이 사용되기 때문.
```

```python
# 두개의 변수를 10, 20 받아서 합계를 구하고 화면에 출력
# 두개의 변수를 10씩 증가하여 합계를 구하고 화면에 출력
# 두개의 변수를 20씩 증가하여 합계를 구하고 화면에 출력
# ...
# 두개의 변수를 50씩 증가하여 합계를 구하고 화면에 출력

a = 10
b = 20

c = a + b
print(c)

a = a + 10
b = b + 10
c = a + b
print(c)

# 이런식으로 직접 작성하여 출력할 수 있지만
# 반복문을 사용하면 더욱 짮게 코드 작성 가능
```

```python
def add(a,b):
    c = a + b
    return c

var1 = 10
var2 = 20

for i in range(0,6): # [0,1,2,3,4,5]
    result = add(var1 + i*10, var2 + i*10)
    print(result)
```

```python
# 데이터프레임의 값 연산

list1 = list([['허준호','남자',24,183],
              ['이가원','여자',23,162],
              ['배규민','남자',21,179],
              ['고고림','남자',28,182],
              ['이새봄','여자',26,160],
              ['이루리','여자',24,163],
              ['오다현','여자',24,157]])
col_names = ['이름','성별','나이','키']
df = pd.DataFrame(list1, columns=col_names)


# 성별 평균 키
mean_by_gender = df.groupby(by=['성별'], as_index=False)['키'].mean( )
mean_by_gender.rename(columns={'키':'평균 키'}, inplace=True)

# 성별 키의 표준편차
std_by_gender = df.groupby(by=['성별'], as_index=False)['키'].std( )
std_by_gender.rename(columns={'키':'키의 표준편차'}, inplace=True)

new_df = pd.merge(mean_by_gender, std_by_gender)
new_df
```
