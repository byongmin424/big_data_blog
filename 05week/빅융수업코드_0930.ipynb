```python
# for in [] 이용해서 0~10 까지 출력 (print)
for i in range(0,11):
  print(i)
```

```python
#for in range()  0부터 10까지의 2의배수 출력
for i in range(0,11,2):
  print(i)
```

```python
#2단 출력 : for in range():
for i in range(1,10):
  print("2 * ", i,"= ", 2*i)
```

```python
#구구단
#for i in range() => 단
#  for j in range() => * j
for i in range(2, 10):
  for j in range(1,10):
    print(i, " * ", j ," = ", i*j)
```

```python
# 두개의 변수 10,20 를 받아서 합계를 구하고-> 화면에 출력
# 두개의 변수에 10을 증가해서 합게를 구하고 -> 화면에 출력
# 두개의 변수에 20을 증가해서 합계를 구하고 -> 화면에 출력
#...두개의 변수에 50을 증가해서 합계 ->화면에 출력
a= 10
b =20

c= a + b
print(c)

a= a+10
b = b+10
c = a +b

print(c)

a = a+50
b = b+50
print(c)
```

```python
#두개의 변수를 받아 합계 출력하는 사용자 함수 생성
def add(a,b):
  c = a+b
  return c

var1=10
var2 =20
# result=add(var1,var2)
# print(result)

# result=add(var1+10, var2+10)
# print(result)

# result=add(var1+20, var2 +20)
# print(result)

for i in range(0,3):
  result=add(var1 + (i*10) ,var2 + (i*10) )
  print(result)
```
