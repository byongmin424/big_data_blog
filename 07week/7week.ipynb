```python
# csv 파일 받아오기
df_csv = pd.read_csv('C:/Users/SAMSUNG/Desktop/big_data_blog/07week/pythonBook.csv',sep=',',encoding='cp949')
df_csv
```

```python
# 데이터 크기
df_csv.shape
```

```python
# 앞 5개 데이터 출력
df_csv.head(5)
```

```python
# 뒤 5개 데이터 출력
df_csv.tail(5)
```

```python
# 데이터 타입
df_csv.info()
```

```python
# 기술 통계
df_csv.describe()
```

```python
# 중복 삭제 _ 출판사
unique_data_csv = df_csv['출판사'].unique()
unique_data_csv
```

```python
# 데이터 수정

# 1. ',' 없애기 _ 가격에서
df_csv['가격'] = df_csv['가격'].str.replace(',','') # ',' -> '' 수정 = 없애는 작업
# 2. 문자형 -> 숫자형으로 변환
df_csv['가격'] = df_csv['가격'].astype('int')

# 데이터 타입 확인
df_csv.info()
```

```python
# 기술 통계
df_csv.describe()
```

```python
# 출간일 순으로 정렬 후 출력 _ sort_values
df_csv.sort_values(by=['출판사','출간일'],ascending=[True,True]).head()
# 오름차순: True, 내림차순: False
```

```python
# 조건식 조회_ (10000 <= x <= 20000) 가격인 책 출력
df_csv[((df_csv['가격']) >= 10000) & ((df_csv['가격']) <= 20000)].head(10)
```

```python
# 특정 문자열 포함하는 문자열 데이터 조회 _ 파이썬
df_csv[df_csv['책이름'].str.contains('파이썬')]
```

```python
# 열 생성 _ ['할인가'] => 가격 * 0.9
df_csv['할인가'] = df_csv['가격'] * 0.9
df_csv.head()
```

```python
# 열 삭제 _ 할인가
df_csv.drop('할인가',axis=1, inplace= True) #inplace 안쓸거면 assign 해줘야해
df_csv.head()
```
