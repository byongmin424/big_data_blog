```python
import pandas as pd
```

```python
df=pd.read_csv('/content/pythonBook.csv',sep=',', encoding='cp949')
df
```

```python
df['가격']=df['가격'].str.replace(',','')
df
```

```python
df['가격']=df['가격'].astype(int)
df.info()
```

```python
df.describe()
```

```python
df.sort_values(by=['출판사','출간일'],ascending=[True,False] )
```

```python
df[  (df['가격']>=10000) & (df['가격']<=20000) ]
```

```python
df[df['책이름'].str.contains('파이썬')]
```

```python
df['할인가'] = df['가격']*0.9
df
```

```python
#집계함수
# groupby() : 옵션 => 1. by => 집계할 항목, 2. 집계함수 => min(), max(), mean(), count()

# 출판사별 서적 의 최대가격 ( 제일 비싼가격)
df.groupby(by='출판사')['가격'].max()
```
