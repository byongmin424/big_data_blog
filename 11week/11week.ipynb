# 9장 표본 평균의 분포

## 통계적 추정
- 모집단 분포(population distribution) : 어떤 통계적 실험 결과인 모집단의 자료가 가지는 확률분포
- 표본 분포(sampling distribution) : 모집단에서 크기 n인 표본을 반복하여 선정할 때 얻어지는 통계량의 확률분포

## 확률표본
- 어떤 모집단으로부터 독립적으로 반복해서 표본을 추출하는 것으로 각각의 관측값들은 서로 독립이며 동일한 분포를 가짐

## 통계량
- 통계적 추론(statistical inference) : 표본의 특성값인 통계량을 이용하여 모집단의 특성을 나타내는 모수를 추론(추측)하는 통계방법
- 통계량(statistic) : 표본평균(sample mean)과 표본분산(sample variance)가 전형적인 통계량이다.

## 표본평균의 분포의 개념
- 표본평균(x-bar)의 분포
    - 통계량은 확률변수이므로 일정한 분포를 갖음
    - 1,2,3,4 로 구성된 크기가 N=4인 모집단으로부터 복원추출의 방법으로 크기가 n-2인 모든 가능한 표본을 추출하고 이를 확률변수 X(i=1,2)라고 하며, X1, X2는 크기가 n=2인 확률표본이 됨

- 표본평균의 분포 : 모분산이 알려진 정규모집단

## 표준오차(Standard Error)
- 표본평균분포의 평균을 이용해 모집단 평균을 추정하는 것을 통계적 추정이라 한다. 표준오차는 이러한 통계적 추정의 정확성을 나타내준다.
- 표본 평균분포의 표준편차가 작을수록 표본평균은 그만큼 모집단 평균을 정확히 추정하고 있음을 의미하므로, 표본평균분포의 표준편차를 '표준오차'라고 부른다.

# 10장 통계적 추정
## 추정의 종류
- 표본자료를 이용하여 모수의 참값이라고 추정하는 하나의 값을 결정하는 점추정(point estimation)과 모수의 참값이 포함되어 있으리라고 추정되는 구간을 결정하는 구간추정(interval estimation)이 있음

## 점추정(point estimation)
- 불편성(unbiasedness) : 추정량의 기대값이 모수와 일치해야 함.
- 효율성(efficiency) : 추정의 error(MSE_mean Squared Error)가 적은것을 효율이 높다고 함. 추정을 여러 번 반복했을 때, 모수와 추정치 차이의 제곱에 대한 평균
- 일치정(consistency) : 표본의 크기가 커질수록 추정량의 오차가 점점 줄어들어, 표본이 매우 클 때는 추정하고자 하는 모수와 추정량이 동일하게 되는 성질
- 추정량(estimator) : 모집단의 특성을 추정하는데 사용되는 표본통계량
- 불편추정량(unbiased estimator) : 추정량의 기대치가 모수와 일치되는 추정량

## 구간추정(interval estimation)
- 점추정의 경우 직관적으로 이해가 가능하고, 계산이 쉽고, 간단하지만 바람직한 추정량을 이용한다 할지라도 그 결과가 얼마나 정확한지 판단하기 어려움
- 따라서 모수의 참값이 포함되어 있으리라는 확신의 정도를 나타내는 척도인 신뢰계수(confidence coefficient)를 가지고 모수의 참값이 포함되어 있으리라고 추정되는 구간을 설정하는 것이 바람직함
- 이러한 추정을 구간추정 또는 신뢰구간추정이라고 함
- 구간추정의 기본개념
    - 신뢰구간 : 구간추정을 통해 얻어지는 특정 구간을 의미
    - 신뢰수준 : 모수의 참값을 포함할 확률(또는 비율)을 의미. 신뢰계수라고도 함.
    - 신뢰한계, 신뢰하한, 신뢰상한
